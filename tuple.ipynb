```python
# Creating an empty Tuple
print("Empty Tuple: ")
Tuple1 = ()
print(Tuple1)
```

```python
# Creating a Tuple with the use of string and numbers
print("\nTuple with the use of Strings andnumbers: ")
Tuple1 = ('Hello', 3, 'Python', 5)
print(Tuple1)
```

```python
# Creating a Tuple with the use of list
print("\nTuple using List: ")
list1 = [1, 2, 4, 5, 6]
print(tuple(list1))
```

```python
# Creating a Tuple with the use of built-in tuple() function

print("\nTuple with the use of function: ")
Tuple1 = tuple('Data Science')
```

```python
# Creating a Set of numbers
print("\nSet: ")
s = {1, 2, 4, 5, 6}
print(s)
```

```python
# Creating a Set using built-in set()
print("\nBlank Set: ")
set1 = set()
print(set1)
```

```python
# Creating a Set with the use of a String
print("\nSet with the use of String: ")
set1 = set("CreationOfSetUsingString")
print(set1)
```

```python
# Creating a Set with a both numbers and strings in a list
print("\nSet with the use of numbers and strings")
set1 = set([1, 2, 'Apple', 4, 'Book', 6,])
print(set1)
```

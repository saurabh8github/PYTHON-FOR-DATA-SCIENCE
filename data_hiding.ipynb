```python
#python program to define abstract class 
from abc import ABC
class Polygon(ABC):
  def sides(self):
    pass
    #normal method
  def a(self):
    print("\nthis is general method.\n")
class Triangle(Polygon):
  def sides(self):
    print("Triangle has 3 sides")
class pentagon(Polygon):
  def sides(self):
    print("pentagon has 5 sides")
class Hexagon(Polygon):
  def sides(self):
    print("hexagon has 6 sides")
class square(Polygon):
  def sides(self):
    print("square has 4 sides")
t=Triangle()
t.sides()
t.a

s=square()
s.sides()

p=pentagon()
p.sides()

h=Hexagon()
h.sides()

```

```python
#creating a base class named univesity:
class University:
  def __init__(self):
    print("constructor of the base class")
    self.univ="MIT"
  def display(self):
    print("The university name is :",self.univ)
class course(University):
  def __init__(self):
    print("consturctot of the child class 1 of class university")
class Branch(University):
  def __init__(self):
    print("constructor of the child class 2 of class university")
    University.__init__(self)
    self.course="cse"
  def display(self):
    print("the course name is :{self.name}")
class student(course,Branch):
  def __init__(self):
    print("constructor of child class of course and branch is called")
    self.name="tony"
    Branch.__init__(self)
    course.__init__(self)
  def display(self):
    print("the name of the student is :{self.name}")
    Branch.display(self)
    course.display(self)
ob=student()
print()
ob.display()

```

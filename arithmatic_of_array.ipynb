```python
#arithmatic operation in numpy
import numpy as np
ar1=np.array([10,11,12,13,14])
ar=np.array([4,6,7,8,8])
new=np.add(ar1,ar)
print(new)
```

```python
#subtraction of array
ar1=np.array([10,11,12,13,14])
ar=np.array([4,6,7,8,3])
new=np.subtract(ar1,ar)
print(new)
```

```python
#divide of the array
ar1=np.array([10,11,12,13,14])
ar=np.array([4,6,7,8,3])
new=np.divide(ar1,ar)
print(new)
```

```python
#modulas of array
ar1=np.array([10,11,12,13,14])
ar=np.array([4,6,7,8,3])
new=np.mod(ar1,ar)
print(new)
```

```python
#multiplication of

ar1=np.array([10,11,12,13,14])
ar=np.array([4,6,7,8,8])
new=np.multiply(ar1,ar)
print(new)
```

```python
#statistical function
ar1=np.array([10,11,12,13,14,5,6,7,2,3])
print(np.amin(ar1),np.amax(ar1))
```

```python
ar1=np.array([50.3,30.2,40.2,60.5,70.7,80.9,99.9])
print(np.percentile(ar1,70))
```

```python
#sorting of one d array
import numpy as np
a=np.array([5,6,73,2,76342,2,46,3,1])
b=np.sort(a)
print(b)
```

```python
#sorting of 2d array
a=np.array([[3,45,67,1,32],[33,445,5,32,7]])
b=np.sort(a)
print(b)
```

```python
#character array sorting
a=np.array(['a','k','z','l'])
b=np.sort(a)
print(b)
```

```python
a=np.array(['johan','rakhi','hai','chhutti','do'])
b=np.sort(a)
print(a)
```

```python
#sorting of one d arrray
import numpy as np
a=np.array([32,3,4,5,23])
print('original array is:',a)
print('sorted array is:',np.sort(a))
```

```python
#descending order numpy array
a=np.array([32,3,4,5,23])
a[::-1].sort()
print(a)
```

```python
#sorting in 2d numpy array
a=np.array([[3,45,67,1,32],[33,445,5,32,7]])
print("with None\n",np.sort(a,axis=None))
print("\n")
print("with axis 1 collumn \n",np.sort(a,axis=1))
print("\n")
print("with axis 0 row\n",np.sort(a,axis=0))
```

```python
#3d array sorting with None ,1,0
new_arr = np.array([[ 78,  23,  41,  66],
              [ 109,  167,  41,  28],
              [ 187, 22, 76, 88]])
b = new_arr.reshape(3, 2, 2)
print("b=",b)
print("\n")
print("By axis b=",np.sort(b,axis=None))
print("\n")
print("By 0 b=",np.sort(b,axis=0))
print("\n")
print("By 1 b=",np.sort(b,axis=1))
```

```python
#aplying sorting algorithms
array=[5,6,73,2,76342,2,46,3,1]
a=np.sort(array,kind='slectionsort')
print(a)
b=np.sort(array,kind='mergesort')
print("\n")
print(b)
c=np.sort(array,kind='quicksort')
print("\n")
print(c)
d=np.sort(array,kind='heapsort')
print("\n")
print(d)
```
